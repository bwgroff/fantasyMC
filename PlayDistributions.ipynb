```python
from pymc import beta_like, gamma_like, lognormal_like

def show_beta(alpha, beta):
    x = np.linspace(0,1,100)
    y = np.exp([float(beta_like(xi, alpha,beta)) for xi in x])
    plt.plot(y)
    
def show_gamma(alpha, beta):
    x = np.linspace(0,1,100)
    y = np.exp([float(gamma_like(xi, alpha,beta)) for xi in x])
    plt.plot(y)
    
def show_ln(mu, tau):
    x = np.linspace(0,1,100)
    y = np.exp([float(lognormal_like(xi, mu, tau)) for xi in x])
    plt.plot(y)
```

```python
def get_plays(starting_yardline, down, turnovers, touchdown, safety, run=True, incomplete=False, db=db):
    q = nfldb.Query(db)
    q.game(season_type="Regular")
    q.play(down=down)
    q.play(yardline=nfldb.FieldPosition(starting_yardline))
    q.play(fumbles_tot=int(turnovers))
    q.play(passing_int=int(turnovers))
    q.play(passing_incmp=int(incomplete))
    q.play(rushing_tds=int(touchdown) * int(run))
    q.play(passing_tds=int(touchdown) * (1-int(run)))
    q.play(rushing_att=int(run))
    q.play(passing_att=1 - int(run))
    return q.as_plays()
```

```python
def get_play_hist(starting_yardline_set, down, turnovers, 
                  touchdown, safety, run=True, incomplete=False, 
                  normalize=False, db=db):
    all_plays = []
    for yl in starting_yardline_set:
        plays = get_plays(yl, down, turnovers, touchdown, safety, run, incomplete, db)
        all_plays += plays
    if run:
        position_dist = Counter([x.rushing_yds for x in all_plays])
    else:
        position_dist = Counter([x.passing_yds for x in all_plays])
    position_pairs = position_dist.items()
    position_pairs.sort(key=lambda x:x[0])
    if normalize:
        num_plays = len(all_plays)
        normalized_pairs = [(x[0],x[1] / float(num_plays)) for x in position_pairs]
        return normalized_pairs
    else:
        return position_pairs
```

```python
results = json.load(open("resultsfull.json"))

team_levels = defaultdict(dict)
for year in ["2009", "2010", "2011", "2012", "2013"]:
    for quality in ["GOOD", "BAD", "MEDIOCRE"]:
        for team in results[quality][year]:
            team_levels[team][year] = quality
```

```python
def get_downsets(game):
    plays = game.plays
    plays.sort(key=lambda x: x.play_id)
    downsets = [[plays[0]]]
    this_downset = []
    team_with_pos = plays[1].pos_team
    for play in plays[1:]:
        if play.pos_team == "UNK": continue
        this_downset.append(play)  # .play_id
        turnover = (play.defense_frec > 0 or play.defense_int > 0)
        first_down = play.first_down
        kick = (play.kicking_tot + play.punting_tot + play.kicking_fgmissed > 0)
        if turnover or first_down or kick:
            downsets.append(this_downset)
            this_downset = []
        team_with_pos = play.pos_team
    return downsets
```

```python
bin_width = 2 # in [2, 4, 5, 10, 20, 25, 50]
bins = 100 / bin_width

def start_state(downset, home):
    pos = str(bin_width * int(downset[0].yardline._offset / bin_width))
    pos_team = downset[0].pos_team
    if pos_team == home:
        team = "H "
    else:
        team = "A "
    return team + pos

def end_state(downset, home):
    return start_state(downset, home)

def ds_stats(downset):
    return downset  # [[(pp.player_id, pp.play_id) for pp in play.play_players] for play in downset]

def state_index(state):
    return bins/2 + (int(state.split(" ")[1]))/bin_width
```

```python
footballMC = defaultdict(lambda: defaultdict(lambda: defaultdict(lambda: defaultdict(list))))
sametransition_counts = defaultdict(lambda: defaultdict(lambda: np.zeros([bins, bins])))  
difftransition_counts = defaultdict(lambda: defaultdict(lambda: np.zeros([bins, bins])))  
# defaultdict(lambda: defaultdict(int))))
master_trans_mat = defaultdict(lambda: defaultdict(int))
all_trans = defaultdict(lambda: defaultdict(lambda: np.zeros([bins, 2*bins])))  

q = nfldb.Query(db)
q.game(season_type="Regular")

for game in q.as_games():
    if game.season_year not in [2009, 2010, 2011, 2012, 2013]:
        continue

    home = game.home_team
    away = game.away_team
    year = game.season_year
    home_qual = team_levels[home][str(year)]
    away_qual = team_levels[away][str(year)]
    
    downsets = get_downsets(game)
    transitions = [(start_state(ds, home), end_state(downsets[i+1], home), 
                    ds_stats(ds)) for i, ds in enumerate(downsets[:-1])]
    transitions.append((start_state(downsets[-1], home), "Game Over", ds_stats(downsets[-1])))
    
    for transition in transitions:
        if transition[0][0] == "H":
            team1_qual = home_qual
            team2_qual = away_qual
        else:
            team1_qual = away_qual
            team2_qual = home_qual
# #         footballMC[team1_qual][team2_qual][transition[0]][transition[1]].append(transition[2])
#         master_trans_mat[transition[0]][transition[1]] += 1
        if transition[1] == "Game Over":
            continue
        index1 = state_index(transition[0])
        index2 = state_index(transition[1])
        if transition[0][0] == transition[1][0]:
            all_trans[team1_qual][team2_qual][index1, index2] += 1
        else:
            all_trans[team1_qual][team2_qual][index1, index2+bins] += 1
```

```python
gvb = normalize(all_trans["GOOD"]["BAD"], norm="l1")
bvg = normalize(all_trans["BAD"]["GOOD"], norm="l1")
plt.figure(7, figsize=(15,80))
for i in range(bins):
    plt.subplot(bins,2,2*i+1)
    plt.plot(gvb[i], c="red")
    plt.plot(bvg[i], c="blue")
    plt.ylim([0,.8])
    plt.subplot(bins,2,2*i+2)
    plt.plot(gvb[i] - bvg[i])
    plt.plot([0] * 2 * bins, c="gray")
    plt.ylim([-.15,.15])
```
